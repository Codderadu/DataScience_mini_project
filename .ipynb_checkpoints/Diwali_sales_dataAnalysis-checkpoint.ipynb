```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("Diwali Sales Data.csv",encoding="unicode_escape")
data
```

```python
data.shape
```

```python
data[data.duplicated()]
```

```python
#drop unrelated/blank 
data.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
pd.isnull(data).sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data["Amount"]=data["Amount"].astype(int)
```

```python
data["Amount"].dtype
```

```python
data[["Age","Orders","Amount"]].describe()
```

```python
data.head()
```

```python
ax=sns.countplot(x="Gender",data=data)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
x=data.groupby("Gender",as_index=False)["Amount"].sum()
x
```

```python
sns.barplot(x='Gender',y="Amount",data=x)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

```python
data
```

```python
ax=sns.countplot(data=data,x="Age Group",hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
# Total Amount vs Age Group
data.head()
```

```python
ax=data.groupby("Age Group",as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sns.barplot(x="Age Group",y="Amount",data=ax)
```

```python
# total number of orders from top 10 states
data.head()
```

```python
orders=data.groupby("State",as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x="State",y="Orders",data=orders)
```

```python
# total amount/sales from top 10 states

sales_state = data.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

```python
# Product Category
data.head()
```

```python
sns.countplot(data=data,x="Product_Category")
```

```python
ax=data.groupby("Product_Category",as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=ax,x="Product_Category",y="Amount")
```

```python
data.head()
```

```python
ax=data.groupby("Product_ID",as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x="Product_ID",y="Amount",data=ax)
```

```python
data.groupby("Product_ID",as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10).plot(kind="bar")
```

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))
data.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

## Conclusion:
Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
